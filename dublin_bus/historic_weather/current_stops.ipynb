```python
import requests
import json
import csv
import datetime
import pandas as pd
```

```python
def call_api(url):
    """Calls API and returns info from that"""

    req = requests.get(url)
    return req

def write_file (data):
    """Retrieves information from Dublin Bikes API and stores as JSON"""

    req_text= data.text
    json_parsed=json.loads(req_text)
    return json_parsed
```

```python
url="https://data.smartdublin.ie/cgi-bin/rtpi/busstopinformation?stopid&format=json"
call_api(url)
data = call_api(url)
json_parsed=write_file(data)
```

```python
i=0
stop_dict=dict() 
bus_stops=dict()
while i < length_stop:
    first = list[i]
    stop_id=first['displaystopid']
    lat=first['latitude']
    long=first['longitude']
    short_name=first['shortname']
    long_name=first['fullname']
    more=first['operators'][0]
    name=more['name']
    routes=more['routes']
    if name == 'bac':
        bac_stops=stop_id
        bac_routes=routes
        bus_stops[bac_stops]=lat, long, long_name
        stop_dict[bac_stops]=bac_routes
        #bus_stops.extend([stop_id, lat, long, long_name])
        print (stop_dict)
    i+=1
```
