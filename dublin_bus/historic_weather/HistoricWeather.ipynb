```python
#Import the required packages
import pandas as pd
import requests
import json
%matplotlib inline
%config IPCompleter.greedy=True

#Display all columns in tables which were being left out before
#https://stackoverflow.com/questions/11707586/python-pandas-how-to-widen-output-display-to-see-more-columns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 154 )
```

```python
# Reading from a csv file, into a data frame
df = pd.read_csv("hly3723.csv")
```

```python
df.head()
```

```python
df['date'] =  pd.to_datetime(df['date'])
```

```python
mask = (df['date'] > '01/01/2016') & (df['date'] <= '31/12/2017')
df_mask = (df.loc[mask])

df_mask
```

```python
df1 = df_mask.drop(['rain', 'wddir', 'clht', 'ind', 'ind.1', 'ind.2', 'ind.3', 'ind.4', 'wetb', 'dewpt', 'vappr', 'w', 'sun', 'vis'], axis=1)
```

```python
df1.rename(columns={'rhum': 'humidity', 'msl': 'pressure', 'wdsp': 'wind_speed', 'wddir': 'wind_direction', 'ww': 'weather_description', 'clht': 'cloud_ceiling_height', 'clamt': 'cloud_amount'}, inplace=True)
```

```python
df1
```

```python
df1.dtypes
```

```python
df1[['wind_speed']] = df1[['wind_speed']].astype(float)
```

```python
# changing wind speed to m/s to match open weather api
df1['wind_speed'] = df1['wind_speed'] * 0.51444444444
```

```python
pd.value_counts(df1['cloud_amount'].values, sort=False)
```

```python
# changing cloud amount to % to match open weather api
df1['cloud_amount'] = df1 ['cloud_amount'] * 12.5
```

```python
df1
```

```python
#Creating bins 
bins = [0, 10, 12, 19, 29, 39, 49, 59, 69, 79, 89, 94, 99]

groupNames = ['clear', 'fog', 'clear', 'precipitation/showers in past hour', 'duststorm/sandstorm/snowstorm', 'fog', 'drizzle', 'rain', 'solid precipitation not in showers', 'showery precipitation', 'thunderstorm in past hour', 'thunderstorm']
array = df1['weather_description']
df1['weather_description_groups']= pd.cut(array, bins=bins, labels=groupNames)
#Creating bins based on descriptive statistics
bins = [0, 10, 12, 19, 29, 39, 49, 59, 69, 79, 89, 94, 99]

array = df1['weather_description']
df1['weather_description_groups']= pd.cut(array, bins=bins, labels=groupNames)
```
